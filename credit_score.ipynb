```python
import pandas as pd
data = pd.read_csv("/content/drive/MyDrive/datasets/train.csv")
```

```python
print(data.isnull().sum())
```

```python
data["Credit_Mix"] = data["Credit_Mix"].map({"Standard": 1, "Good": 2, "Bad": 0})
```

```python
import plotly.express as px
fig = px.box(data, x="Occupation", color="Credit_Score", title="Credit Scores by Occupation")
fig.show()
```

```python
fig = px.box(data, x="Credit_Score", y="Annual_Income", color="Credit_Score", title="Credit Scores by Annual Income")
fig.show()
```

```python
from sklearn.model_selection import train_test_split
x = data[["Annual_Income", "Monthly_Inhand_Salary", "Num_Bank_Accounts", "Num_Credit_Card",
          "Interest_Rate", "Num_of_Loan", "Delay_from_due_date", "Num_of_Delayed_Payment",
          "Credit_Mix", "Outstanding_Debt", "Credit_History_Age", "Monthly_Balance"]]
y = data["Credit_Score"]
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(xtrain, ytrain)
```

```python
# Define input features
features = [[19114.12, 1824.843333, 2, 2, 9, 2, 12, 3, 2, 250, 200, 310]]

# Make a prediction
predicted = model.predict(features)

# Display the prediction
print("Predicted Credit Score: ", predicted)

```
